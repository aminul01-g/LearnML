# 5.Extracting needed Columns for 429 pages

```python
# Initialize an empty list to store DataFrames
dataframes = []

for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
    
    # Check if the response is successful
    if response.status_code == 200:
        temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        dataframes.append(temp_df)  # Append the DataFrame to the list
    else:
        print(f"Failed to retrieve data for page {i}: {response.status_code}")

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dataframes, ignore_index=True)
```
