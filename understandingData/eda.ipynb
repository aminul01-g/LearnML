## Univariate Analysis

### 1.Catagorical Data

#### a.Countplot

```python
sns.countplot(df['Sex'])
#df['Survived'].value_counts().plot(kind='bar')
```

#### b.PieChart

```python
df['Survived'].value_counts().plot(kind="pie",autopct = '%.2f')
```

### 2.Numerical Data

#### a.Histogram

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins= 5)
```

#### b.Distplot

```python
sns.distplot(df['Age'])
```

#### c.Boxplot

```python
sns.boxplot(df['Age'])
```
